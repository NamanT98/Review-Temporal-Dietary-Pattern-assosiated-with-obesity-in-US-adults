```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
pd.pandas.set_option("Display.max_column",None)
```

```python
df=pd.read_sas("DR1IFF_C.XPT")
df["Year"]=["2003-04"]*len(df)
df
```

```python
df3=pd.read_sas("DR1IFF_D.XPT")
df3["Year"]=["2005-06"]*len(df3)
df3.head()
```

```python
df_comb=df.append(df3)
df_comb
```

```python
df_comb=df_comb.loc[df_comb.SEQN>1,["Year","SEQN","DR1IGRMS","DR1_020","DR1IKCAL","DR1ISUGR","DR1ITFAT","DR1ICHOL","DR1IALCO"]]
```

```python
demo=pd.read_sas("DEMO_C.XPT")
demo2=pd.read_sas("DEMO_D.XPT")
demo_com=demo.append(demo2)
demo_com
```

```python
demo_com=demo_com.loc[demo_com.SEQN>1,["SEQN","RIAGENDR","RIDRETH1","INDFMPIR","RIDAGEYR"]]
demo_com
```

```python
bm=pd.read_sas("BMX_C.XPT")
bm2=pd.read_sas("BMX_D.XPT")
bm_com=bm.append(bm2)
bm_com
```

```python
bm_com=bm_com.loc[bm2.SEQN>1,["SEQN","BMXWT","BMXBMI","BMXWAIST"]]
bm_com
```

```python
pa=pd.read_sas("PAQ_C.XPT")
pa2=pd.read_sas("PAQ_D.XPT")
pa_com=pa.append(pa2)
pa_com
```

```python
pa_com=pa_com.loc[pa_com.SEQN>1,["SEQN","PAD080"]]
pa_com
```

```python
preg=pd.read_sas("UC_C.XPT")
preg2=pd.read_sas("UCPREG_D.XPT")
preg_com=preg.append(preg2)
preg_com
```

```python
df_comb.dropna(inplace=True)
demo_com.dropna(inplace=True)
bm_com.dropna(inplace=True)
preg_com.dropna(inplace=True)
pa_com.dropna(inplace=True)
```

```python
print(df_comb.isnull().sum())
print(demo_com.isnull().sum())
print(bm_com.isnull().sum())
print(preg_com.isnull().sum())
print(pa_com.isnull().sum())
```

```python
demo_l=list(demo_com.SEQN.unique())
bm_l=list(bm_com.SEQN.unique())
preg_l=list(preg_com.SEQN.unique())
pa_l=list(pa_com.SEQN.unique())
index={"seqn":[],"demo":[],"bm":[],"pa":[]}
for seqn in df_comb.SEQN.unique():
    if seqn in demo_l:
        if seqn in bm_l:
            if seqn in pa_l:
                    index["seqn"].append(seqn)
                    index["demo"].append(demo_com.iloc[demo_l.index(seqn),1:])
                    index["bm"].append(bm_com.iloc[bm_l.index(seqn),1:])
                    index["pa"].append(pa_com.iloc[pa_l.index(seqn),1:])
```

```python
val=list(df_comb.SEQN.values)
row=[]
i=0
while i<len(val):
    if val[i] in index["seqn"]:
        row.append(i)
    i+=1
```
